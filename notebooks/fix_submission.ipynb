# Fix Submission CSV Format
Converts from `(file_base, label)` → `(id, genre)` with zero-padded sequential IDs.

```python
import pandas as pd

# ── Config ────────────────────────────────────────────────────────────────────
INPUT_PATH  = "../submissions/submission.csv"   # your current file
OUTPUT_PATH = "../submissions/submission_fixed.csv"
ID_DIGITS   = 4                  # zero-padding width  →  0001, 0002 …
# ──────────────────────────────────────────────────────────────────────────────

df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df)} rows")
print(df.head())
```

```python
# Validate expected columns exist
assert 'file_base' in df.columns, "Expected column 'file_base' not found"
assert 'label'     in df.columns, "Expected column 'label' not found"

# Build the new dataframe
fixed = pd.DataFrame({
    'id':    [str(i).zfill(ID_DIGITS) for i in range(1, len(df) + 1)],
    'genre': df['label'].values
})

print("\nReformatted preview:")
fixed.head(10)
```

```python
# Quick sanity checks
print(f"Total rows  : {len(fixed)}")
print(f"ID range    : {fixed['id'].iloc[0]}  →  {fixed['id'].iloc[-1]}")
print(f"Genre counts:\n{fixed['genre'].value_counts()}")
```

```python
fixed.to_csv(OUTPUT_PATH, index=False)
print(f"Saved → {OUTPUT_PATH}")
```
